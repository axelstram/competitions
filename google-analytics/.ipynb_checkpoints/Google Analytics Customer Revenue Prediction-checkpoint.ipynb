```python
def RMSE(actual, pred):
    return (np.mean((actual - pred) ** 2)) **.5
```

```python
def group_target_by_fields(df, target, fields = None):
    if fields is None:
        fields = df.columns
    
    target_mean = df[target].mean()
    target_std = df[target].std()
    
    d = {}
    
    for f in fields:
        gp = df.groupby(f)[target]
        mean = gp.mean()
        std  = gp.std()
        
        if pd.DataFrame(mean).isnull().values.any():
            mean = mean.fillna(target_mean)
            
        if pd.DataFrame(std).isnull().values.any():
            std = std.fillna(target_std)
            
        df[f + '_' + target + '_avg'] = df[f].map(mean)
        df[f + '_' + target + '_std'] = df[f].map(std)
        
        d[f + '_avg'] = mean
        d[f + '_std'] = std
        
    return d
```

```python
def apply_group_by_dict(df, d, fields_to_group_by, target):
    for f in fields_to_group_by:
        mean = d[f + '_avg']
        std = d[f + '_std']

        df[f + '_' + target + '_avg'] = df[f].map(mean)
        df[f + '_' + target + '_std'] = df[f].map(std)
```

```python
def get_folds(df=None, n_splits=5):
    """Returns dataframe indices corresponding to Visitors Group KFold"""
    # Get sorted unique visitors
    unique_vis = np.array(sorted(df['fullVisitorId'].unique()))

    # Get folds
    folds = GroupKFold(n_splits=n_splits)
    fold_ids = []
    ids = np.arange(df.shape[0])
    for trn_vis, val_vis in folds.split(X=unique_vis, y=unique_vis, groups=unique_vis):
        fold_ids.append(
            [
                ids[df['fullVisitorId'].isin(unique_vis[trn_vis])],
                ids[df['fullVisitorId'].isin(unique_vis[val_vis])]
            ]
        )

    return fold_ids
```

```python
num_folds = 5
folds = get_folds(df=X, n_splits=num_folds)

excluded_features = [
    'date', 'fullVisitorId', 'sessionId', 'transactionRevenue', 
    'visitId', 'visitStartTime'
]

train_features = [_f for _f in X.columns if _f not in excluded_features]

assert(len(X.columns) == len(X_test.columns))

importances = pd.DataFrame()
oof_reg_preds = np.zeros(X.shape[0])
sub_reg_preds = np.zeros(X_test.shape[0])
for fold_, (trn_, val_) in enumerate(folds):
    trn_x, trn_y = X[train_features].iloc[trn_], y_reg.iloc[trn_]
    val_x, val_y = X[train_features].iloc[val_], y_reg.iloc[val_]
    
    params = {}
    params["device"] = "gpu"
    params["metric"] = "rmse"
    
    reg = lgb.LGBMRegressor(**params,
                max_depth= 19,
                num_leaves= 1491,
                min_child_weight=147.96151086680055,
                subsample= 0.6844278953034877,
                subsample_freq= 3,
                lambda_l1= 14.554169985906551,
                lambda_l2= 39.86756353974596,
                feature_fraction= 0.7199869681784914,
                learning_rate= 0.03202860673214656,
                min_data_per_leaf= 608,
                min_gain_to_split= 0.2538208632983941,
                n_estimators=3000,
                random_state=1
                )
    reg.fit(
        trn_x, np.log1p(trn_y),
        eval_set=[(val_x, np.log1p(val_y))],
        early_stopping_rounds=50,
        verbose=100,
        eval_metric='rmse'
    )
    imp_df = pd.DataFrame()
    imp_df['feature'] = train_features
    imp_df['gain'] = reg.booster_.feature_importance(importance_type='gain')
    
    imp_df['fold'] = fold_ + 1
    importances = pd.concat([importances, imp_df], axis=0, sort=False)
    
    oof_reg_preds[val_] = reg.predict(val_x, num_iteration=reg.best_iteration_)
    oof_reg_preds[oof_reg_preds < 0] = 0
    _preds = reg.predict(X_test[train_features], num_iteration=reg.best_iteration_)
    _preds[_preds < 0] = 0
    sub_reg_preds += np.expm1(_preds) / len(folds)
    
mean_squared_error(np.log1p(y_reg), oof_reg_preds) ** .5
```
